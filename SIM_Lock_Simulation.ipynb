```python
import time
import random
import string

SIM_LOCKED = True  # Default sim lock status is Locked
SECRET_KEY = None

# Simulated database for storing username, password, and SIM numbers
user_database = {
    "Alice": {
        "password": "password123",
        "otp": None
    },
    "Bob": {
        "password": "qwerty456",
        "otp": None
    },
    "Charlie": {
        "password": "pass987",
        "otp": None
    }
}

# Simulated database for storing SIM numbers and secret keys
sim_database = {}

# Function to generate a random SIM number
def generate_sim_number():
    return ''.join(random.choice(string.digits) for _ in range(10))

# Function to generate a random secret key
def generate_secret_key():
    characters = string.ascii_letters + string.digits + string.punctuation
    return ''.join(random.choice(characters) for _ in range(16))

# Function to generate a random OTP
def generate_otp():
    return ''.join(random.choice(string.digits) for _ in range(6))

# Function to lock the SIM and generate a secret key
def lock_sim():
    global SIM_LOCKED, SECRET_KEY
    if not SIM_LOCKED:
        SIM_LOCKED = True
        SECRET_KEY = generate_secret_key()
        print("SIM locked successfully. \n")
    else:
        print("SIM is already locked.\n")

# Function to unlock the SIM
def unlock_sim():
    global SIM_LOCKED
    if SIM_LOCKED:
        SIM_LOCKED = False
        print("SIM unlocked successfully.\n")
    else:
        print("SIM is already unlocked.\n")

# Function to prompt security questions and validate answers
def validate_security_questions():
    answers = {
        "What is your favorite color?": "Blue",
        "What is your pet's name?": "Simba",
        "What city were you born in?": "Bangalore"
    }
    for question, answer in answers.items():
        user_answer = input(question + " ")
        if user_answer != answer:
            return False
    return True

# Function to validate the username and password with the simulated database
def validate_credentials(username, password):
    if username in user_database and user_database[username]["password"] == password:
        return True
    return False

# Function to send OTP to the user
def send_otp(username):
    otp = generate_otp()
    user_database[username]["otp"] = otp
    print("OTP sent to your number XXXXX 98653:", otp)

# Function to validate the OTP
def validate_otp(username, otp):
    stored_otp = user_database[username]["otp"]
    if stored_otp and stored_otp == otp:
        return True
    return False

# Function to simulate SIM lock and security questions
def simulate_sim_lock():
    print("Starting SIM lock simulation...")
    print("Current SIM lock status:", "Locked \n" if SIM_LOCKED else "Unlocked\n")

    # Prompt user for username and password
    username = input("Enter your username: ")
    password = input("Enter your password: ")

    # Validate credentials with the simulated database
    if validate_credentials(username, password):
        print("Username and password validated successfully. \n")

        # Send OTP to the user
        send_otp(username)
        entered_otp = input("Enter the OTP received: ")

        # Validate OTP
        if validate_otp(username, entered_otp):
            print("OTP validation successful.")
            unlock_sim()  # Unlock the SIM
            time.sleep(2)  # Simulating some delay

            # Validate security questions if SIM is unlocked
            if not SIM_LOCKED:
                print("- SECURITY QUESTIONS-")
                print("Answer all security questions to swap SIM \n")
                if validate_security_questions():
                    print("Security questions successfully answered.\n")
                    old_sim_number = generate_sim_number()
                    new_sim_number = generate_sim_number()
                    sim_database[old_sim_number] = new_sim_number
                    swap_sim(old_sim_number, new_sim_number)
                else:
                    print("Invalid security question answers. SIM remains locked.\n")

            lock_sim()  # Lock the SIM again
        else:
            print("Invalid OTP. Exiting simulation.")
    else:
        print("Invalid username or password. Exiting simulation.")

    # print("Simulation completed.\n")
    # print("Final SIM lock status:", "Locked \n" if SIM_LOCKED else "Unlocked \n")


    # Ask user if they want to freeze their new SIM with a secret key
    if SIM_LOCKED:
        choice = input("Do you want to freeze your new SIM with a secret key? (yes/no) ")
        if choice.lower() == "yes":
            freeze_sim(new_sim_number)
        else:
          print("Thank you. Contact T-Mobile Customer Support 1-800-937-8997 for further queries.")

# Function to simulate SIM swapping
def swap_sim(old_sim_number, new_sim_number):
    print("SIM swapped successfully.")
    print("Old SIM number:", old_sim_number)
    print("New SIM number:", new_sim_number)

# Function to freeze the new SIM with a secret key
def freeze_sim(sim_number):
    global SIM_LOCKED, SECRET_KEY
    if not SIM_LOCKED:
        SIM_LOCKED = True
        SECRET_KEY = generate_secret_key()
    secret_key = generate_secret_key()
    sim_database[sim_number] = secret_key
    print("New SIM frozen successfully with secret key:", secret_key)
    print("Make note of this secret key for future reference, contact T-Mobile Customer Support 1-800-937-8997 for further queries.")
    SIM_LOCKED = True  # Unlock the SIM
    print("Simulation completed.\n")
    print("Final SIM lock status:", "Locked \n" if SIM_LOCKED else "Unlocked \n")

# Run the simulation
simulate_sim_lock()
```
